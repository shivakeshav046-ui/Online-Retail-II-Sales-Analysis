# Data Insert and Cleaning

```python
import pandas as pd 
```

```python
file_path = r"D:\OneDrive\Desktop\Eda project\project_data\online_retail_II.xlsx"

df_2009 = pd.read_excel(file_path, sheet_name="Year 2009-2010")
df_2010 = pd.read_excel(file_path, sheet_name="Year 2010-2011")

concated_data = pd.concat([df_2009, df_2010], ignore_index=True)

data = pd.DataFrame(concated_data)
print("Data Frame Created Successfully")
print(f"shape of the data: {data.shape}")
```

### Removing Duplicates

```python
duplicate_rows = data.duplicated().sum()
if duplicate_rows == 0:
    print("No Duplicates Found")
else :
    data.drop_duplicates(inplace=True) 
    print("Deleted Duplicates")
    print(f"Here is the New shape: {data.shape}")
```

```python
data["Invoice"] = data["Invoice"].astype(str)
```

## Observation
- Dropped rows where **Price = 0** because they are **invalid transactions**

```python
invalid_data = data[data["Price"] == 0]
data.drop(invalid_data.index,inplace=True)
print("--"*50)
print("Dropped unnecessary data")
print(f"shape of the data {data.shape}")
print("description nulls = ",data["Description"].isnull().sum())
print("--"*50)
```

## Observation

- All rows with missing **Description** had **Price = 0**.
- Removing rows where **Price = 0** also removed all missing values in **Description**.

## Observation
